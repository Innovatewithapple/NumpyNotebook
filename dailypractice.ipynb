Series

```python
mys = [1,2,3,4]
ind = ['a','b','c','d']
s = pd.Series(mys,index=ind)
print(s['d'])
```

Create series as key value dict

```python
myd = {"a":"apple","b":"Banana"}
d = pd.Series(myd)
d
```

Suppose you wanna put some key values in series not all:

```python
myd = {"a":1,"b":2,"c":3,'d':4}
d = pd.Series(myd,index=['a','b'])
d
```

```python

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

dc = data['calories']
createindex=[]
for i in range(len(dc)):
  createindex.append(f"day{i+1}")
  print(createindex)

mydataframe = pd.DataFrame(data,index=createindex)
mydataframe
```

concatenate

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr3 = np.concatenate((arr1,arr2))
print(arr3)
```

```python
arrl = np.array([[1,2,3,4]])
arrp = np.array([[8,9,0,11]])
arrt = np.concatenate((arrl,arrp))
print(arrt)
```
