Slicing_____  

```python
#[start:end:step] here is the rule to remember: the end point always gives value of a previous index (5 means 4th index)
print(indArr)
print(indArr[0:1])
```

```python
#Slice elements from index 4 to the end of the array:
print(indArr[4:])
```

```python
# Negetive slicing:
print(indArr[-3:-1])
```

```python
#Return every other element from index 1 to index 5 using STEP:
print(indArr)
print(indArr[1:5:1]) # now here the steps never start counting with 0 so when you set 1 it basicallly counts the same number no nothing to remove
print(indArr[1:5:2])
```

```python
# Return every other element from the entire array:
print(indArr[::2])
```

```python
# Slicing 2D Array:
Two_arr = np.array([
    [1,2,3,4,5,6,20],
    [7,8,9,10,11,12,30]])
print(Two_arr)

#From the second element, slice elements from index 1 to index 4 (not included):
print("slicing:")
print(Two_arr[1,1:-1:2])
```

```python
# From both elements, return index 2:
print(Two_arr[::,2])
```

Copy() VS View()

```python
arr_Original = np.array([1,2,3,4])
arr_copy = arr_Original.copy()
arr_Original[0] = 5
print(arr_Original)
print(arr_copy)
```

```python
arr_view = arr_Original.view()
arr_Original[0] = 0
print(arr_Original)
print(arr_view)
```

```python
arr_v = arr_Original
arr_Original[0] = -1
print(arr_Original)
print(arr_v)
```

```python

arr = np.array([1, 2, 3, 4, 5])

x = arr.copy()
y = arr.view()

print(x.base)
print(y.base)
```

Concatenate

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9])
arr4 = np.array([11,1,2])
arr3 = np.concatenate((arr2,arr1,arr4))
print(arr3)
```

```python
#Two D array concatenate
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[7,8,9],[10,11,12]])
arr3 = np.concatenate((arr1,arr2),axis=0)
print(arr3)
```

Filter

```python
arr = np.array([41, 42, 43, 44])

x = [True, False, True, False]

newarr = arr[x]

print(newarr)
```

```python
arr = np.array([41, 42, 43, 44])

filter_arr = arr > 42

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

filter_arr = arr % 2 == 0

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```
